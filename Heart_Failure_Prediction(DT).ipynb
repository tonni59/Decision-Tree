```python
import pandas as pd
import numpy as np
from sklearn import tree
import pydotplus
from sklearn.model_selection import train_test_split

import matplotlib.pyplot as plt
import matplotlib.image as pltim
```

```python
df = pd.read_csv("/content/drive/MyDrive/DATA MINING & MACHINE LEARNING/heart_failure.csv")
df
```

**Check NUll Value**


```python
df.isnull().values.any()
```

```python
df.describe()
```

**Seperate Feature Columns**


```python
x = df.drop(['DEATH_EVENT'] , axis = 1)
x
```

**Seperate Class Column**


```python
y = df['DEATH_EVENT']
y
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size = 0.3 , random_state=1)
```

Create object of Decision Tree Classifier, Fit our Train dtata into our model


```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier()
dtc.fit(xtrain,ytrain)
```

**Prediction and Evalution**


```python
predictions = dtc.predict(xtest)
```

```python
predictions
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(ytest,predictions))
```

```python
print(confusion_matrix(ytest,predictions))
```

```python
dtc.score(xtest,ytest)
```
